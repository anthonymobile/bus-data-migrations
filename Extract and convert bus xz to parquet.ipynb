```python
# get file list
import glob
filelist = glob.glob('./nyc/transitcenter/data/*.xz')
filelist.sort()
```

```python
# extract each into a df
# dump df to parquet
import pandas as pd

for f in filelist:
    df = pd.read_csv(f)
    df.to_parquet(f'{f.split(".csv")[0]}.parquet')
    print (f'extracted:{f}'')
```
